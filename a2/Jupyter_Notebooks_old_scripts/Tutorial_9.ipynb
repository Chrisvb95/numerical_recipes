```python
from scipy.misc import imread
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
```

```python
wolf = Image.open('wolf.jpg')
wolf_fft = np.fft.fft(wolf)
tile = np.tile(wolf,(2,2))

plt.imshow(wolf,cmap ='gray')
plt.show()
plt.imshow(tile,cmap ='gray')
plt.show()
plt.imshow(np.log(np.abs(wolf_fft)),cmap ='gray')
plt.show()
```

```python
wolf_fft = np.fft.fftshift(wolf_fft)
plt.imshow(np.log(np.abs(wolf_fft)),cmap ='gray')
plt.show()
```

```python
median = np.median(wolf_fft)
stddev = np.std(wolf_fft)
print(median,stddev)
```

```python
copy_wolf_fft = np.copy(wolf_fft)
indices = np.where(wolf_fft > median + 1.5*stddev)
wolf_fft[indices] = median + np.random.normal(median,stddev/2.,len(indices[0])) 
plt.imshow(np.log(np.abs(wolf_fft)))
plt.show()
```

```python
wolf_fft = np.fft.ifftshift(wolf_fft)
wolf_fft = np.real(np.fft.ifft(wolf_fft))
plt.imshow(wolf_fft,cmap ='gray')
plt.show()
```
