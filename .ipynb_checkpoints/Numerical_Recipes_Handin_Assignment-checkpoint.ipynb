```python
import numpy as np
import matplotlib.pyplot as plt
import sys

seed = 42
print('Seed value used:',seed)
```

```python
# Functions for first task.
def poisson_distribution(mean,k):
# Returns probability for given k and mean
    fact = np.float64(1) 
    for i in range(round(k)):
        fact = round(fact*(k-i),6) # MAKE A FIX FOR LARGE NUMBERS
        
    print('Factorial:',fact)
    if fact >= sys.maxsize:
        print('Factorial exceeds float max:',fact)
        return
    return (mean**(k)*np.exp(-mean))/fact
#end poisson_distribution

def LCG_gen(x):
# Linear Congruenial generator
    a,c,m = 2**32,1664525,1013904223
    return np.int64((a*x+c)%m)
#end LCG_gen

def XOR_shift(x):
# XOR-shift generator
    a1,a2,a3 = 21,35,4
    x = x ^ x >> a1
    x = x ^ x << a2 
    x = x ^ x >> a3 
    return x
#end XOR_shift

def rand_num(l,seed,min=0,max=1):
# Combination of XOR_shift and LCG_gen
    x = list([np.int64(seed)])
    for i in range(l):
        x[i] = LCG_gen(x[i])
        x[i] = XOR_shift(x[i])
        x[i] = LCG_gen(x[i])
        x[i] = XOR_shift(x[i])
        x[i] = XOR_shift(x[i])
        x.append(x[i])
    x = np.array(x)/sys.maxsize
    return min+(x[:-1]*(max-min))
# MAKE CLASS
```

```python
# Poisson distribution
a1 = [[1,0],[5,10],[3,20],[2.6,40],[101,200]]
for i in range(len(a1)):
    print(poisson_distribution(a1[i][0],a1[i][1]))

```

```python
# Functions used for differentiation
def central_diff(f,h,x):
# Calculates the central difference
    return (f(x+h)-f(x-h))/(2*h) 
#end central_diff()

def ridders_differentiator(f,h,x,m):
# Differentiates using Ridder's method
    def D_calc(i,j,d,D,D_new):
    # Calculates D
        return (d**(2*(j+1)) * D[j-1] - D_new[j-1]) / (d**(2*(j+1))- 1)
    #end D_calc()
    D = np.zeros((m,200))
    d = 2
    for i in range(m):
        D_new = D      
        for j in range(i+1):    
            if j == 0:
                D_new[j] = central_diff(f,h,x)
            else:
                D_new[j] = D_calc(i,j,d,D,D_new)  
        D = D_new    
        h = h/d
        #print(D)
    return D[m-1]
    # Could still look into iterating until a certain accuracy is reached                                                      
#end ridders_differentiator()

def S_calc(i,j,S,S_new):
        #return (4**(j+1) * S[j-1]) - S_new[j-1] / (4**(j+1)- 1)
        return (S_new[j-1]+(1/(4**j-1))*(S_new[j-1]-S[j-1])
def h_calc(n,a,b):
        return 1/(2**n)*(b-a) 
def Sn0(n,a,b,f,S):
    sum = 0
    if n == 0:
        h = h_calc(1,a,b)
        return h*(b-a)
    else:
        h = h_calc(n,a,b)
        for i in range(1,int(2**(n-1))+1):
            sum += f(x+(2*i-1)*h)
        #print(sum)
        #print(0.5*S+h*sum)
        return 0.5*S+h*sum
    
def romber_int(f,x,a,b):
    #Calculate S(0,0)
    n = 0  
    S = np.zeros((1,len(x)))
    #S = np.array([h_calc(n,a,b)*(f(a)+f(b))])
    for i in range(5):
        n += 1
        S_new = np.zeros((len(S)+1,len(x)))
        S_new[0] = Sn0(n,a,b,f,S[0])
        #print(S_new[0])
        #print(f(x))
        for i in range(1,len(S_new)):
            S_new[i] = S_calc(n,i,S,S_new)
            #print(S_new)
        S = S_new
    print(S)
    return S

def trapezoid(f,h,x):
# Calculates the central difference
    return h*(f(x+h)+f(x-h)) 
#end central_diff()

def rombergs_integrator(f,h,x,m):
# Integrates using the romberg technique   
    def S_calc(i,j,d,S,S_new):
        return (4**(j+1) * S[j-1]) - S_new[j-1] / (4**(j+1)- 1)
    def h_calc(n,a,b):
        return 1/(2**n)*(b-a) 
    S = np.zeros((m,200))
    d = 2
    n = 1
    for i in range(m):
        n *= 2
        S_new = S      
        for j in range(i+1):    
            if j == 0:
                
                S_new[j] = trapezoid(f,h,x)
            else:
                S_new[j] = S_calc(i,j,d,S,S_new)  
        S = S_new    
        h = 1/n
    #print(S)
    return S[m-1]
#end rombergs_integrator()



x = np.linspace(0,2*np.pi,200)

plt.plot(x,x**2*np.sin(x))
plt.title('function')
plt.show()
    
D = ridders_differentiator(lambda x:x**2*np.sin(x),0.001,x,5)

plt.plot(x,D)
plt.title('Ridders dif')
plt.show()
                                                       
S = rombergs_integrator(lambda x:x**2*np.sin(x),1,x,5)
#print(sum(S))
plt.plot(x,S)
plt.title('Romb int')
plt.show()

S = romber_int(lambda x:x**2*np.sin(x),x,1,5)
#print(S)
#plt.plot(x,S)
#plt.title('Romb int')
#plt.show()
                                                       
#Functions used for integration
```
