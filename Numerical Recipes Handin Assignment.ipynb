```python
import numpy as np
import matplotlib.pyplot as plt

seed = 42
print('Seed value used:',seed)
```

```python
def poisson_distribution(k,mean):
# Returns probability for given k and mean
    fact = 1 
    for i in range(round(k)):
        fact = fact*(k-i)
    return (mean**(k)*np.exp(-mean))/fact
#end poisson_distribution

def LCG_gen(x):
# Linear Congruenial generator - returns l integers
    a,c,m = 2**32,1664525,1013904223
    return int((a*x+c)%m)
    #end LCG_gen

def XOR_shift(x):
# XOR-shift generator - returns l integers
    a1,a2,a3 = 21,35,4
    x = x ^ x >> a1
    x = x ^ x << a2 
    x = x ^ x >> a3 
    return x
#end XOR_shift

def rand_num(l,seed):
# Combination of XOR_shift and LCG_gen
    x = [seed]
    for i in range(l-1):
        x[i] = LCG_gen(x[i])
        x[i] = XOR_shift(x[i])
        x[i] = LCG_gen(x[i])
        while x[i] > 1:
            x[i] = x[i]/10.
        x.append(x[i])
    return x 
```
